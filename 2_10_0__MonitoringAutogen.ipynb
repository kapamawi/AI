````python
from autogen import ConversableAgent, AssistantAgent, UserProxyAgent
from typing import Annotated, Literal
from autogen import ConversableAgent, register_function
import autogen
import agentops
import os
from dotenv import load_dotenv
from IPython.core.error import  StdinNotImplementedError
from google.colab import userdata
import sqlite3
import pandas as pd
````

````python
class CFG:
  model = "gpt-4o-mini"
````

````python
OPENAI_API_KEY = userdata.get('openaivision')
````

````python
config_list = [
    {
        "model": CFG.model,
        "api_key": OPENAI_API_KEY,
        "tags": ["mathagent-example", "tool"],
    }
]
````

# Test 1

````python

logging_session_id = autogen.runtime_logging.start(config={"dbname": "logs.db"})
print("Logging session ID: " + str(logging_session_id))


assistant = AssistantAgent(name="assistant", llm_config={"config_list": config_list})
user_proxy = UserProxyAgent(
    name="user_proxy",
    code_execution_config=False,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "TERMINATE" in msg["content"],
)

user_proxy.initiate_chat(
    assistant, message= "Why did the chicken cross the road? "
)


autogen.runtime_logging.stop()
````

Ten kod konfiguruje i uruchamia prostą konwersację między dwoma agentami używając biblioteki autogen:

1. Rozpoczyna sesję logowania:
```python

```
Utworzy to bazę danych SQLite o nazwie "logs.db" do zapisywania informacji o przebiegu rozmowy.

2. Tworzy dwa agenty:
- `AssistantAgent` - agent AI zdolny do prowadzenia rozmowy
- `UserProxyAgent` - agent symulujący użytkownika z następującymi ustawieniami:
  - Wykonywanie kodu jest wyłączone (`code_execution_config=False`)
  - Nie będzie pytał o input od człowieka (`human_input_mode="NEVER"`)
  - Zakończy rozmowę gdy napotka słowo "TERMINATE" w wiadomości

3. Rozpoczyna rozmowę między agentami pytaniem "Why did the chicken cross the road?"

4. Na końcu zatrzymuje logowanie sesji wywołaniem `autogen.runtime_logging.stop()`
